```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('bank').getOrCreate()
bank = spark.read.csv('/home/ubuntu/Loan Defaulter.csv', header=True, inferSchema=True)
```

```python
print(bank.columns)
print(bank.show)
```

```python
bank.printSchema()
```

```python
bank.createOrReplaceTempView("bank")
bank = spark.sql("SELECT * FROM bank WHERE TARGET = 1 ")
import pandas as pd
pd.DataFrame(bank.take(bank.count()), columns=bank.columns)
```

```python
bank.createOrReplaceTempView("bank")
bank = spark.sql("SELECT * FROM bank WHERE TARGET = 0 and CODE_GENDER = 'F' ")
import pandas as pd
pd.DataFrame(bank.take(bank.count()), columns=bank.columns)
```
