```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('bank').getOrCreate()
bank = spark.read.csv('/home/ubuntu/Loan Defaulter.csv', header=True, inferSchema=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Convert sex to an array using Numpy and plot it using pyplot. 
TARGETArr = np.array(bank.select('TARGET').collect())
plt.hist(TARGETArr)
plt.show()
```
